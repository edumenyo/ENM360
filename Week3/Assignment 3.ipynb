## Problem 1:

Here we are looking for the probability of --- given the knowledge our friend picked up the mango as our fruit ($\Pr(F=M)$) --- that mango having come from bag 2 ($\Pr(B=B_2)$). Overall, this is represented by $\Pr(B=B_2 \mid F=M)$.

We can then use Bayes rule to use $\Pr(B=B_2)$ as our prior: $$\frac{\Pr(F=M \mid B=B_2) \ * \ \Pr(B=B_2)}{\Pr(F=M)}$$

Then we can apply the Sum rule to the denominator ($\Pr(F=M) \text{ becomes} \sum_B \big(\Pr(F=M \mid B) \ * \ \Pr(B) \big)$).

Simplified, our total equation is $$ \frac{\Pr(F=M \mid B=B_2) \ * \ \Pr(B=B_2)} {\Pr(F=M \mid B=B_1)\Pr(B=B_1) \ + \ \Pr(F=M \mid B=B_2)\Pr(B=B_2)} $$

For the numerator, we are given that the likelihood of selecting Box 2 is 0.4 ($\Pr(B=B_2)$) and that the chance of selecting a mango from Box 2 is $4/8 = 0.5$ ($\Pr(F=M \mid B=B_2)$). Thus, the expression simplifies to $(0.4)(0.5)=(0.2)$.

Similarly, for the denominator, we are told that the likelihoods of selecting from Box 1 ($\Pr(B=B_1) = 0.6$) and of selecting a mango from Box 1 ($\Pr(F=M \mid B=B_1) = \frac{2}{3}$). Thus, the expression on the denominator becomes ($(0.5)(0.4) + (\frac{2}{3})(0.6) = 0.6$).

Combining the numerator and denominator, our overall answer is $\Pr(B=B_2 \mid F=M) = \frac{0.2}{0.6} = 0.\overline{33}$.


## Problem 2:
#### a) Solving for the expectation and variance of $\Pr(x)$ 
* expectation of $\Pr(x)$:

We are given $\Pr(x) = \alpha \Pr_1(x) \ + \ (1 - \alpha) \Pr_2(x)$. Notation-wise, let's consider the expectation of $\Pr_1(x)$ as $\mathbb{E_1}[x]$ and $\Pr_2(x)$ as $\mathbb{E_2}[x]$. Furthermore, we know that the expectation of $\Pr[x]$ --- $\mathbb{E}(x)$ --- is defined as $\int_{-\infty}^\infty (\Pr(x) * x) \: dx$. Thus we proceed as follows:

We can plug in for $\Pr(x)$ in our expectation equation: $\mathbb{E}[x] = \int_{-\infty}^\infty \big( [\alpha \Pr_1(x) \ + \ (1 - \alpha) \Pr_2(x) ] * x \big) \: dx$

We can separate out this integral into $\int_{-\infty}^\infty \alpha \Pr_1(x)*x \: dx \ + \ \int_{-\infty}^\infty (1 - \alpha)\Pr_2(x)*x \: dx$. After factoring out the constants $\alpha$ and $(1-\alpha)$, we have $\alpha \int_{-\infty}^\infty \Pr_1(x) *x \: dx \ + \ (1 - \alpha) \int_{-\infty}^\infty \Pr_2(x)*x \: dx$. This simplifies to $\mathbb{E}[x] = \alpha \mu_1 + (1-\alpha)\mu_2 = \mu$ given the definition of $\mathbb{E}[x]$.


* variance of $\Pr(x)$:

Similarly to as occurred above, we will use the definition of variance ($\mathbb{Var}[x] = \mathbb{E}[(x- \mathbb{E}[x])^2] = \mathbb{E}[x^2] - \mathbb{E}[x]^2 = \mu^2$).

We then substitute for $\mathbb{E}[x]$, which immediately resolves to $\mathbb{Var}[x] = \mathbb{E}[x^2] \ + \ (\alpha \mu_1 + (1-\alpha)\mu_2)$. To simplify $\mathbb{E}[x^2]$ requires more work, however. Given our 2 normal distributions, it is not too difficult to realize we will have $\mathbb{E}[x^2] = \alpha \int_{-\infty}^\infty \Pr_1(x) *x^2\: dx \ + \ (1 - \alpha) \int_{-\infty}^\infty  \Pr_2(x) *x^2\: dx$. To solve these integrals, let us consider our earlier equation $\mathbb{E}[x^2] - \mathbb{E}[x]^2 = \mu^2$ --- once we plug in $\sigma^2$ for $\mathbb{E}[x]^2$, we can solve to $\mathbb{E}[x^2] = \sigma^2 + \mu^2$.

Now given that general form for $\mathbb{E}[x^2]$, we can apply it to each of the integrals (each of which takes the expectation of ) so that we have  $\mathbb{E}[x^2] = \alpha \int_{-\infty}^\infty \Pr_1(x) *x^2\: dx \ + \ (1 - \alpha) \int_{-\infty}^\infty  \Pr_2(x) *x^2\: dx \\ \qquad = \alpha (\sigma_1^2 + \mu_1^2) + (1 - \alpha)(\sigma_2^2 + \mu_2^2) $.

Finally, we are ready to solve for $\mathbb{Var}[x] = \mathbb{E}[x^2] - \mathbb{E}[x]^2$: we now have $$ \mathbb{Var}[x] = \big(\alpha (\sigma_1^2 + \mu_1^2) + (1 - \alpha)(\sigma_2^2 + \mu_2^2) \big) -  \big(\alpha \mu_1 + (1-\alpha)\mu_2  \big)$$

INSTRUCTIONS: Consider a bivariate Gaussian distribution
$\Pr(x,y)=\mathcal{N}(\: [\begin{smallmatrix}0 \\ 1 \end{smallmatrix}], \:
            [\begin{smallmatrix} 5 & −2 \\ −2 & 7 \end{smallmatrix}] \: )$.

First, use the properties of the Gaussian distribution to derive the density for the marginal $\Pr(x)$
and the conditional $\Pr(x \mid y=1)$. Then write a Python script to generate 1000 samples from the $\Pr(x,y)$, $\Pr(x)$, and $\Pr(x \mid y=1)$ using Numpy’s built-in function `numpy.random.multivariate_normal`. Finally use SciPy’s built-in Gaussian kernel density estimator to fit a probability density for each set of samples and plot it against the densities you derived analytically.

SOLUTION: Given the properties from lecture, we can find $\Pr(x)$ as $\mathcal{N}(x \mid 0, 5) $. 

The density for $\Pr(x \mid y=1)$ can be also be found using the properties of the Gaussian: it's $\mathcal{N}(\mu_{x \mid y = 1}, \Sigma_{x \mid y = 1})$, where $\mu_{x \mid y = 1} = [0] + (-2)(1/7)(1 - 1) = 0$ and $\Sigma_{x \ \mid \ y =  1} = [5] - (-2)(1/7)(-2) = 4.43$


```python
import numpy as np
import sympy as sp
from scipy import stats as spstats
import matplotlib.pyplot as plt

# so we need to find a way to compose a mixture of Pr1(x) and Pr2(x)

class GaussianDistributionPlotter:
    def __init__(self):
        self.i = 0
        self.mu = np.array([0, 1])
        self.sigma = np.array([(5, -2), (-2, 7)])
        self.muX = np.array([0, 0])
        self.sigmaX = np.array([(5, 0), (0, 0)])
        
        self.muXGivenYIs1 = self.mu[0] + self.sigma[0][1]*(1 / self.sigma[1][1])*(1 - self.mu[1])
        self.muXGivenYIs1 = np.array([self.muXGivenYIs1, 0])
        
        self.sigmaXGivenYIs1 = self.sigma[0][0] - self.sigma[0][1]*(1 / self.sigma[1][1])*self.sigma[1][0]
        self.sigmaXGivenYIs1 = np.array([(self.sigmaXGivenYIs1, 0),(0, 0)])
        
         # print("mu: {}\nsigma:\n{}".format(mu, sigma))
         # print("muX:\n{}".format(muX))
         # print("sigmaX:\n{}".format(sigmaX))
         # print("muXGivenYIs1:\n{}".format(muXGivenYIs1))
         # print("sigmaXGivenYIs1:\n{}".format(sigmaXGivenYIs1))
        
    def __plot(self, plot=None, data=None):
        plotChoices = ["XY", "X", "X|Y=1"]
        
        if data is None or plot is None or plot not in plotChoices:
            raise ValueError()
             
        plt.figure(self.i)
        self.i += 1
            
        if plot == "XY":
            samplesXY = data[0]
            X, Y, Z = data[1]
            plt.hist2d(samplesXY.T[0], samplesXY.T[1], bins=150)
            plt.contour(X, Y, Z, linewidths=2.5)
            plt.colorbar()
            plt.xlabel('$y_1$')
            plt.ylabel('$y_2$')
            plt.title("Density of $Pr(x,y)$ calculated by KDE (contour) vs analysis (histogram)")
            
        elif plot == "X":             
            kdePDFOfX = data[0]
            analysisPDFOfX = data[1]
            Xpoints = data[2]
             
            plt.plot(Xpoints, kdePDFOfX, '-o', label="KDE")
            plt.plot(Xpoints, analysisPDFOfX, '--', label="Analysis")
            plt.xlabel('$x$')
            plt.ylabel('$Pr(x)$')
            plt.title("Density of $Pr(x)$ calculated by analysis vs. by KDE")
            plt.legend()
             
        elif plot == "X|Y=1":             
            kdePDFOfXGivenYIs1 = data[0]
            analysisPDFOfXGivenYIs1 = data[1]
            XGivenYIs1points = data[2]
             
            plt.plot(XGivenYIs1points, kdePDFOfXGivenYIs1, '-o', label="KDE")
            plt.plot(XGivenYIs1points, analysisPDFOfXGivenYIs1, '--', label="Analysis")
            plt.xlabel('$x$')
            plt.ylabel('$Pr(x \mid y=1)$')
            plt.plot()
            plt.title("Density of $Pr(x \mid y=1)$ calculated by analysis vs. by KDE")
            plt.legend()
        
    def computeAndPlotSamples(self):
         # returns a (1000, n) array of samples --- 1000 samples for each distribution, n Gaussian functions
         samplesXY = np.random.multivariate_normal(self.mu, self.sigma, 1000)
         samplesX = np.random.multivariate_normal(self.muX, self.sigmaX, 1000)
         samplesXGivenYIs1 = np.random.multivariate_normal(self.muXGivenYIs1, self.sigmaXGivenYIs1, 1000)
         # print("samplesXY size: {}\nsamplesX size: {}\nsamplesXGivenYIs1 size: {}".format(samplesXY.shape,
         #                                                                                  samplesX.shape,
         #                                                                                  samplesXGivenYIs1.shape))

         XYpoints = np.zeros((2, 125))
         XYpoints[0] = np.linspace(-15, 15, 125)
         XYpoints[1] = np.linspace(-15, 15, 125)
         Xpoints = np.linspace(-25, 25, 125)
         XGivenYIs1points = np.linspace(-25, 25, 125)

         tempnorm = spstats.norm(self.muX[0] , np.sqrt(self.sigmaX[0][0]))
         analysisPDFOfX = np.array([tempnorm.pdf(x) for x in Xpoints])

         tempnorm = spstats.norm(self.muXGivenYIs1[0], np.sqrt(self.sigmaXGivenYIs1[0][0]))
         analysisPDFOfXGivenYIs1 = np.array([tempnorm.pdf(x) for x in XGivenYIs1points])

         # computing and evaluate the kernel density estimates
         kdeXY = spstats.gaussian_kde(samplesXY.T)
         kdeX = spstats.gaussian_kde(samplesX[:, 0].T)
         kdePDFOfX = kdeX.evaluate(points=Xpoints)
         kdeXGivenYIs1 = spstats.gaussian_kde(samplesXGivenYIs1[:, 0].T)
         kdePDFOfXGivenYIs1 = kdeXGivenYIs1.evaluate(points=XGivenYIs1points)
        
         X, Y = np.meshgrid(XYpoints[0], XYpoints[1])
         positions = np.vstack([Y.ravel(), X.ravel()])
         Z = np.reshape(kdeXY(positions).T, X.shape)
                
         self.__plot(plot="XY", data=(samplesXY, (X, Y, Z)))
         self.__plot(plot="X", data=(kdePDFOfX, analysisPDFOfX, Xpoints))
         self.__plot(plot="X|Y=1", data=(kdePDFOfXGivenYIs1, 
                                       analysisPDFOfXGivenYIs1, 
                                       XGivenYIs1points))
         
if __name__ == "__main__":
    gaussDP = GaussianDistributionPlotter()
    gaussDP.computeAndPlotSamples()
```

## Problem 4:
INSTRUCTIONS: Consider a bivariate Gaussian random variable X

distributed according to the joint density
$\Pr(x_1,x_2)=\mathcal{N}(\: [\begin{smallmatrix} 0 \\ 0 \end{smallmatrix}], \:
[\begin{smallmatrix} 1.0 & 0.9 \\ 0.9 & 1.0 \end{smallmatrix}] \:)$.

Now consider another random variable Y
with components given by a non-linear transformation of X, namely

$$(y_1,y_2)=(αx_1,x_2/α+β(x_1^2+α^2)),$$
with α=1.15 and β=0.8.

Write a Python code to generate 5000 realizations of X
, and use them to generate 5000 samples of Y. Plot the samples of Y along with a [Gaussian kernel density estimator](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gaussian_kde.html). Also compute and report the the empirical mean of Y.

-----------------------------
SOLUTION:


```python
import numpy as np
import sympy as sp
from scipy import stats as spstats
import matplotlib.pyplot as plt

class TransformationOfGaussianRandomVariable:
    '''generates samples over the distribution N([0; 0], [1.0 0.9; 0.9 1.0])'''

    def generateXSamples(self, numSamples):
        sampX = np.random.multivariate_normal(self.mean, self.variance, numSamples)
        return sampX
    
    def generateYSamples(self, samplesX, numSamples):
        sampY = np.copy(samplesX)
        sampY[0] = self.alphaConst * samplesX[0]
        sampY[1] = (samplesX[1] / self.alphaConst) + self.betaConst*(np.square(samplesX[0]) + self.alphaConst**2)
        
        return sampY

    def computeEmpiricalMean(self, samples=None):
        if samples is None:
            raise ValueError()
            
        size = 0
        total = 0
        for sampRow in samples:
            size += sampRow.size
            total += np.sum(sampRow)
#            print("length: {}, new total: {}".format(size, total))
            
        return total / size
    
    def plotKDE(self, samples, plotEmpiricalMean=True):
        kernel = spstats.gaussian_kde(samples.T)
        y_star = np.array([(np.linspace(-2.1, 2.1, 100)), 
                            (np.linspace(-2.1, 2.1, 100))])

        plt.figure(self.i)
        self.i += 1
        
        X, Y = np.meshgrid(y_star[0], y_star[1])
        positions = np.vstack([Y.ravel(), X.ravel()])
        Z = np.reshape(kernel(positions).T, X.shape)
        CS = plt.contour(X, Y, Z, linewidths=2.5)
        plt.clabel(CS, inline=1, fontsize=10)
        plt.xlabel('$y_1$')
        plt.ylabel('$y_2$')
        plt.title('PDF plot of Y using kernel density estimate')
        
        if plotEmpiricalMean is True:             
            print("The empirical mean is: {}".format(self.computeEmpiricalMean(samples.T)))      
    
    def __init__(self, mean, variance, alphaConst, betaConst):
        self.i = 0
        self.mean = mean
        self.variance = variance
        self.alphaConst = alphaConst
        self.betaConst = betaConst
        

if __name__ == "__main__":
    
    alphaConst = 1.15
    betaConst = 0.8
    
    mean = np.array([0, 0])
    variance = np.array([(1.0, 0.9), (0.9, 1.0)])
    
    transGRV = (TransformationOfGaussianRandomVariable(mean, variance,
                                                       alphaConst, betaConst))
    
    samplesX = transGRV.generateXSamples(5000)   
    samplesY = transGRV.generateYSamples(samplesX, 5000)
    
    transGRV.plotKDE(samplesY, plotEmpiricalMean=True)
    
```
